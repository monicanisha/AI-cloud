```python
import pandas as pd
data=pd.read_excel(r'top_seeker_searches_all_data.xlsx')
print(data.head(10))
```

2.bar chart

```python
x=data['Company']
y=data['result_count']
plt.bar(x,y,color='white',edgecolor='black',linewidth=2)
plt.title("Tips Dataset")
plt.ylabel('Total_result')
plt.xlabel('seeker')
plt.show()
```

3.

```python
x=data['result_count']
plt.hist(x)
plt.bar(x,y,color='red',edgecolor='blue',linewidth=2)
plt.title("Tips Dataset")
plt.ylabel('Total_result')
plt.xlabel('seeker')
plt.show()
```

4.histogram

```python
x=data['Company']
y=data['result_count']
plt.hist(x)
plt.title("Tips Dataset")
plt.ylabel('Total_result')
plt.xlabel('seeker')
plt.show()
```

```python
x=data['result_count']
plt.hist(x,bins=25,color='red',edgecolor='blue',linestyle='-',alpha=0.5)
plt.title("Tips Dataset")
plt.ylabel('frequency')
plt.xlabel('Total_result')
plt.show()
```

```python
6. scatter plot
```

```python
x=data['Company']
y=data['result_count']
plt.scatter(x,y)
plt.title("Tips Dataset")
plt.ylabel('frequency')
plt.xlabel('Total_result')
plt.show()
```

```python
5. pie chart
```

```python
seekers=['COMPANY','ROLE','FUNCATION','DESIGNSTION','LOCATION']
data=[23,10,45,15,12]
plt.pie(data,labels=seekers)
plt.title("search seeker")
plt.show()
```

```python
seekers=['COMPANY','ROLE','FUNCATION','DESIGNSTION','LOCATION']
data=[23,10,45,15,12]
explode=[0.1,0.5,0,0,0]
colors=("orange","cyan","yellow","grey","green")
plt.pie(data,labels=seekers,explode=explode,autopct='%1.2f%%',colors=colors,shadow=True)
plt.title("search seekers")
plt.show()
```
